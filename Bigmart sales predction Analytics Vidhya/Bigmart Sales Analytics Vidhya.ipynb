## About Regularization

Some predictors get high weightage in comparison to others. Hence the model is overfitting when it comes to these features. It is not the correct thing to happen because the model is not giving due importance to the other features and hence we need to introduce **Regularization**.

Regularization will introduce a parameter called **Alpha**. This is also called as shrinkage parameter. It will shrink the parameters which have very high value and it will normalize it.

There are 2 regularization techniques:
* **L1** (Lasso Regression)
* **L2** (Ridge Regression)

### Lasso

```python
from sklearn.linear_model import Lasso,LassoCV,Ridge,RidgeCV

X = dummy_train.drop('Item_Outlet_Sales',axis = 1)
y = dummy_train.Item_Outlet_Sales

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2,random_state = 42)


#Simple Lasso model to see the effect of trian and test RMSE
lasso = Lasso(alpha = 1.0,normalize = True) #Normalize is done for the model to give equal weightage. Put always.
pred_train = lasso.fit(xtrain,ytrain).predict(xtrain)
pred_test = lasso.fit(xtrain,ytrain).predict(xtest)

rmse(ytrain,pred_train),rmse(ytest,pred_test)
```

```python
plt.figure(figsize = (15,6))
pd.Series(lasso.coef_).sort_values().plot(kind = 'bar')
plt.show()
#As we decrease the alpha value, coefficients are not restricted and most of the variables are still 0.
```

```python
# Function that tells me which predictors are 0s basis alpha values

def lassos(alphas):
    # create a dataframe
    df = pd.DataFrame()
    
    #Names of the Columns/Predictors
    df["Features"] = xtrain.columns
    
    # Loop for all alphas
    for i in alphas:
        lasso = Lasso(alpha=i)
        lasso.fit(xtrain,ytrain)
        col_name = "Alpha = %f" % i
        df[col_name] = lasso.coef_
    return(df)
```

```python
#[0,0.1,0.01,0.001,0.5,0.99,1,2,5,10]
lassos([0,0.1,0.01,0.001,0.5,0.99,1,2,5,10,20,30,50]) #as we introduce more bias, the model simplifies but it will lose the prediction power.
```

```python
#Finding the RMSE and R2 behaviour basis different alphas.
from sklearn.metrics import mean_squared_error,r2_score

rmse_train = []
rmse_test = []
a1 = []
r2 = []
alphas = [0.001,0.09,0.01,0.1,0.5,0.99,1,2,5,10,20,30,50]

for i in alphas:
    lasso = Lasso(alpha = i,normalize = True)
    a1.append(i)
    pred_tr = lasso.fit(xtrain,ytrain).predict(xtrain)
    pred_test = lasso.fit(xtrain,ytrain).predict(xtest)
    rmse_train.append(np.sqrt(mean_squared_error(ytrain,pred_tr)))
    rmse_test.append(np.sqrt(mean_squared_error(ytest,pred_test)))
    r2.append(r2_score(ytest,pred_test))
```

```python
lasso_df = pd.DataFrame({"Alpha":a1,"RMSE_Train":rmse_train,"RMSE_Test":rmse_test,"Rsq":r2})
lasso_df.sort_values(by = "Rsq",ascending = False)
#Inf: As we introduce higher values of alpha, model complexity reduces.
# At the same time, the R2 score reduces and error increases. This is the point where 
# the prediction power gets compromised.
```

```python
plt.figure(figsize = (15,6))
plt.plot(lasso_df.RMSE_Train,label = "training error")
plt.plot(lasso_df.RMSE_Test,label = "test error")
plt.ylabel('RMSE')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
#If test error is higher than training error
# you need to increase the test %age.

#Apply CV to find the same

#max_iter - how many times the model will run...

lassocv = LassoCV(cv=5,normalize=True,max_iter = 1000)

pred_tr = lassocv.fit(xtrain,ytrain).predict(xtrain)
pred_test = lassocv.fit(xtrain,ytrain).predict(xtest)
print("train error: ",(np.sqrt(mean_squared_error(ytrain,pred_tr))))
print("Test Error: ",(np.sqrt(mean_squared_error(ytest,pred_test))))
```
