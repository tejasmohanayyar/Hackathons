### Let us perform some EDA on the data first.

```python
#Let us look at outlier detection and cleaning.
Q1 = new_train.quantile(0.25)
Q3 = new_train.quantile(0.75)
IQR = Q3-Q1
new_data1 = new_train[~((new_train>(Q3+(1.5*IQR))) | (new_train<(Q1-(1.5*IQR)))).any(axis = 1)]
```

### Imputing missing values

```python
new_data1.isnull().sum()[new_data1.isnull().sum()>0]
```

```python
sns.distplot(new_data1['LotFrontage'])
plt.show()
new_data1['LotFrontage'] = new_data1['LotFrontage'].fillna(method = 'ffill')
new_data1['LotFrontage'].isnull().sum()
```

```python
new_data1['MasVnrArea'] = new_data1['MasVnrArea'].fillna(0)
new_data1[new_data1['YearBuilt']==new_data1['GarageYrBlt']].shape
```

### Lets create a Regression Model and check

```python
import statsmodels.api as sm
from statsmodels.api import OLS
from sklearn.model_selection import train_test_split
x = new_data1.drop(['SalePrice'],axis = 1)
x = sm.add_constant(x,has_constant='add')
y = new_data1.SalePrice

#Performing the routine train test split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2,random_state = 2)
model = OLS(ytrain,xtrain).fit()
```

```python
model.params[model.params==0].index
```

```python
import statsmodels.api as sm
from statsmodels.api import OLS
from sklearn.model_selection import train_test_split
x = new_data1.drop(['SalePrice','EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal',],axis = 1)
x = sm.add_constant(x,has_constant='add')
y = new_data1.SalePrice

#Performing the routine train test split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2,random_state = 2)



model = OLS(ytrain,xtrain).fit()
print(model.summary())
```

```python
from sklearn.metrics import mean_squared_error
RMSE = np.sqrt(mean_squared_error(ytest,model.predict(xtest)))
RMSE
```

###### Linear regression seems to be doing quite well in this model. Let us test for:

* Linearity (Rainbow test)
* Normality (Shapiro, Jarque Bera)
* Heteroskedasticity (Bouschepegan)
* Auto-Correlation (Durbin Watson)
* Multi-Collinearity (VIF,Heatmap)

```python
import statsmodels.api as sm
sm.stats.diagnostic.linear_rainbow(model) #The model is linear
```

```python
from scipy.stats import probplot
plt.figure(figsize = (15,6))
probplot(model.resid,plot = plt)
plt.show()
#We see considerable deviation in the data and hence can conclude that this model fails the test of normality.
```

###### Linearity is passed since Durbin Watson between 1.95 and 2.5

###### Normality:

```python
#The above model fails the test of normality.
stats.shapiro(model.resid)
```

###### Heteroskedasticity: Our model has passed the breusch pagan test. The model is not showing heteroskdastic pattern.

* Perform the Bouschepegan test

```python
import statsmodels.stats.api as sm
teststat,pvalbp,ftest,pvalftest = sm.het_breuschpagan(model.resid,xtrain) #endog,exog
pvalbp
```

```python
sns.scatterplot(model.resid,model.fittedvalues)
plt.ylabel('fitted values')
plt.xlabel('Residuals')
plt.show()
#Looking at the graph, we can see some tendencies of heteroskedasticity
```

###### No correlation is present in the residuals as indicated by durbin-watson test.
