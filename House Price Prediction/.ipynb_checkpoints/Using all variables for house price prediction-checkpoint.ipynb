###### Apply linear regression and the 5 checks wrt regression.
* Normality
* Linearity
* Heteroskedasticity
* Auto-Correlation
* Collinearity

```python
import statsmodels.api as sm
from statsmodels.api import OLS
from sklearn.model_selection import train_test_split
x = reg_data.drop(['SalePrice'],axis = 1)
x = sm.add_constant(x,has_constant='add')
y = reg_data.SalePrice

#Performing the routine train test split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2,random_state = 2)
model = OLS(ytrain,xtrain).fit()
print(model.summary())

#Tests of regression
# type your code here
print('\n\n')
import statsmodels.api as sm
from scipy.stats import shapiro
print('Rainbow test of linearity:',sm.stats.diagnostic.linear_rainbow(model)) #Passes test of linearity
print('Shapiro test of normality:',shapiro(model.resid))
import statsmodels.stats.api as sma
teststat,pvalbp,ftest,pvalftest = sma.het_breuschpagan(model.resid,xtrain) #endog,exog
print('breuschpagan test for heteroskedasticity:',pvalbp)
```

```python
from sklearn.metrics import mean_squared_error
RMSEt = np.sqrt(mean_squared_error(ytest,model.predict(xtest)))
RMSEtr = np.sqrt(mean_squared_error(ytrain,model.predict(xtrain)))
 
RMSEt,RMSEtr
```

###### Let us check multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

x = reg_data.drop('SalePrice',axis = 1)
vifs = []
for i in range(len(x.columns)):
    vifs.append(vif(x.values,i))
```

```python
vif_scores = pd.DataFrame({'Features':x.columns,'VIF':vifs})
acceptable_vif = vif_scores[vif_scores['VIF']<10]
acceptable_vif
```

##### Inferences

* Model passes auto-correlation test. (Durbin Watson)
* Model is a valid one since prob(F-stat) is less than 0.05.
* High condition number means multicollinearity exists.
* Model passes test of linearity according to the rainbow test.
* Model fails the Shapiro Wilk test of normality.
* Model fails the breuschpagan test for heteroskedasticity.
* In the above case the model has overfit and features must be further looked into to better generalize the model.
* We will go back to feature engineering and try to look into the data.
* 20 values were shown to have a vif greater than 10. After removing we are left with 15 numerical parameters. We will now try a new model with just these 15 parameters to check for improvement.

```python
import statsmodels.api as sm
from statsmodels.api import OLS
from sklearn.model_selection import train_test_split
x = reg_data[acceptable_vif['Features']]
x = sm.add_constant(x,has_constant='add')
y = reg_data.SalePrice

#Performing the routine train test split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2,random_state = 2)
model = OLS(ytrain,xtrain).fit()
print(model.summary())

#Tests of regression
# type your code here
print('\n\n')
import statsmodels.api as sm
from scipy.stats import shapiro
print('Rainbow test of linearity:',sm.stats.diagnostic.linear_rainbow(model)) #Passes test of linearity
print('Shapiro test of normality:',shapiro(model.resid))
import statsmodels.stats.api as sma
teststat,pvalbp,ftest,pvalftest = sma.het_breuschpagan(model.resid,xtrain) #endog,exog
print('breuschpagan test for heteroskedasticity:',pvalbp)
```

```python
from sklearn.metrics import mean_squared_error
RMSEt = np.sqrt(mean_squared_error(ytest,model.predict(xtest)))
RMSEtr = np.sqrt(mean_squared_error(ytrain,model.predict(xtrain)))
 
RMSEt,RMSEtr
```

###### Our model seems to do very poorly. We can conclude that it is not necessary that a high VIF for a particular column means that it does not contribute to the model. However we have reduced the overfitting problem but the test data seems to have a very high score. This means that we are probably underfitting in the above case. The difference between train and test RMSE has reduced by 6000 points compared to before we removed high multi-collinear columns

### Imputing Ordinal Categorical Features first

```python
ordinal_features = cat_cols[cat_cols[cat_cols=='Gd'].any()[cat_cols[cat_cols=='Gd'].any().values==True].index]
```

```python
# ordinal_features.columns
```

```python
def ordinal_feature_imp(data):
    data.replace({'NA':0},inplace = True)
    data['BsmtExposure'].replace({'No':1,'Mn':2,'Av':3,'Gd':4},inplace = True)
    for i in ['ExterQual','ExterCond','BsmtQual','BsmtCond','HeatingQC','KitchenQual','FireplaceQu','GarageQual','GarageCond','PoolQC']:
        data[i].replace({'Po':0,'Fa':1,'TA':2,'Gd':3,'Ex':4},inplace = True)
    
    return data
```

```python
ordinal_feature_imp(ordinal_features)
```

```python
impu_cat_cols = pd.concat((cat_cols.drop(ordinal_features.columns,axis = 1),ordinal_features),axis = 1)
```

```python
combined_data = pd.concat((num_cols,pd.get_dummies(impu_cat_cols,drop_first = True)),axis = 1)
combined_data.shape
```

```python
combined_data.dtypes[combined_data.dtypes=='O']
```

```python
import statsmodels.api as sm
from statsmodels.api import OLS
from sklearn.model_selection import train_test_split

x = sm.add_constant(combined_data.drop('SalePrice',axis = 1))
y = combined_data.SalePrice

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 1)
mod = OLS(y_train,x_train).fit()
print(mod.summary())
```

```python
from sklearn.metrics import mean_squared_error
RMSE_train = np.sqrt(mean_squared_error(y_train,mod.predict(x_train)))
RMSE_test = np.sqrt(mean_squared_error(y_test,mod.predict(x_test)))
RMSE_train,RMSE_test
```

###### While the RMSE has gone down, we are overfitting on the test set.
